```python
import os

root = "/home/jupyter-228w1a5418/riceproject"
search_terms = ["cloud_removal","eval.py","utils.py","train.py", "predict.py","models_utils.py","pretrained_models","demo.py","config.yml" ,"models/", "data_manager"]

for dirpath, _, filenames in os.walk(root):
    for f in filenames:
        if f.endswith(('.py', '.ipynb', '.yml', '.txt')):
            full_path = os.path.join(dirpath, f)
            try:
                with open(full_path, 'r', errors='ignore') as file:
                    for line_num, line in enumerate(file, start=1):
                        if any(term in line for term in search_terms):
                            print(f"{full_path}:{line_num}: {line.strip()}")
            except:
                pass
```
